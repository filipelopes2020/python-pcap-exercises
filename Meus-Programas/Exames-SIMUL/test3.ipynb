```python
class Gamma:
    def __init__(self, val):
        self.name = val
    def __str__(self):
        return "I am " + self.name + " from Gamma"
    
class Delta:
    def __init__(self, val):
        self.name = val
    def __str__(self):
        return "I am " + self.name + " from Delta"
    
class Beta(Gamma, Delta):
    def __str__(self):
        return super().__str__()
    
class Omega(Beta):
    omega = 3
    pass
    
o = Omega("omega")
print(o)
```

```python
class Omega:
    def __init__(self):
        self.set(3)
    def set(self, x):
        self.x = x//2
    
class Lambda(Omega):
    L = 4
    def __init__(self):
        super().__init__()
    def __str__(self):
        return str(self.x/3)[-1]
    def set(self, x):
        self.x = x//2 + 1
    
L = Lambda()
print(L)

print(L.x)          # 2
print(L.x/3)        # 0.6666666666666666 (a float)
print(str(L.x/3))   # 0.6666666666666666  (a string)
print(str(L.x/3)[-1]) # 6 (last character in string '0.6666666666666666'
print(L)            # 6
```
