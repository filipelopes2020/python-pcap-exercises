```python
class A:
    x = 'x'
    def alarm(self):
        print('a' + self.x)

class B(A):
    def alarm(self):
        super().alarm()

class C(B):
    x = 'y'

C().alarm()   # ay

a = A()
a.alarm()     # ax

b = B()
b.alarm()     # ax, porque self é instância de B, e B não redefine x
```

```python
def contador(maximo):
    for i in range(1, maximo + 1):
        yield i  # Produz i, mas mantém a função "congelada"

gen = contador(5)

t=list(gen)
print(t)  # [1, 2, 3, 4, 5]
```

```python
def sum_list(args, fun):
    z = 0
    for x in args:
        z = z + fun(x)
    return z
    
print([ex//2 for ex in range(5)])
print([ (lambda x: 1 if x > 1 else 0)(x) for x in [0, 0, 1, 1, 2] ])
print(sum_list([ex//2 for ex in range(10)], lambda x: 1 if x>1 else 0))
```

```python
class Robot:
    def t(self):
        print('Robot')
    
class Droid:
    def t(self):
        print('Droid')
    
class ProtoDroid(Droid, Robot):
    def f(self):
        self.t()
        
class FighterDroid(Robot, Droid):
    def f(self):
        self.t()
    
R2D2 = ProtoDroid()
R2D2.f()                    # Droid
    
C3PO = FighterDroid()
C3PO.f()                    # Robot
```

```python
class AlphaDivisionError(ZeroDivisionError):	
    def __init__(self, message):
        ZeroDivisionError.__init__(self, message)
    
def func_div(x,y):
    if y==0:
        raise ZeroDivisionError("Can't divide by zero !")
    elif isinstance(y, str):		
        raise AlphaDivisionError("Can't divide by string !")
    else:		
        return x/y
    
try:
    #print(func_div(4,'a'))
    print(func_div(4,0))
except ZeroDivisionError as e:
    print(str(e),"OOO")                    # Can't divide by string !
except Exception as e:
    print("There is an error !")
else:
    print('All good !')
finally:
    print('End of program !')
```
